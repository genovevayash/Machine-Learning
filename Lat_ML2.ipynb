```python
# Import Library

import pandas as pd
import numpy as np
```

```python
df_final = pd.read_csv('/content/df_final.csv')
```

```python
df_final
```

```python
df_final.info()
```

```python
# ganti nilai untuk fitur score
df_final['kategori'] = np.where(df_final['score_overall'] <= 8.9, "buruk", "baik")
```

```python
# ganti nilai untuk fitur age_group
conditions = [
    (df_final['age_group'] == '18-24' ),
    (df_final['age_group'] == '25-34' ),
    (df_final['age_group'] == '35-44' ),
    (df_final['age_group'] == '45-54' ),
    (df_final['age_group'] == '55+')
]

choices = ['1', '2', '3', '4', '5']

df_final["age"] = np.select(conditions, choices, default='0')

print(df_final)
```

```python
df_final.info()
```

```python
df_final.describe()
```

```python
df_final.columns
```

```python
from sklearn import preprocessing

# label_encoder object knows
# how to understand word labels.
label_encoder = preprocessing.LabelEncoder()

df_final['city'] = label_encoder.fit_transform(df_final['city'])
df_final['country'] = label_encoder.fit_transform(df_final['country'])
df_final['user_gender'] = label_encoder.fit_transform(df_final['user_gender'])
df_final['traveller_type'] = label_encoder.fit_transform(df_final['traveller_type'])
df_final['kategori'] = label_encoder.fit_transform(df_final['kategori'])
df_final['age'] = label_encoder.fit_transform(df_final['age'])
```

```python
df_final
```

```python
# ambil beberapa kolom aja
df_new = df_final[['city', 'country', 'user_gender', 'join_year', 'age', 'kategori']]

print(df_new)
```

```python
df_new.info()
```

```python
df_new.columns
```

```python
df_new['kategori'].value_counts()
```

```python
X = df_new[['city', 'country', 'user_gender',
       'join_year', 'age']]

y = df_new['kategori']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.40, random_state=50)
```

```python
X_train
```

```python
X_test
```

### Modelling Machine Learning

```python
from sklearn.tree import DecisionTreeClassifier

model1 = DecisionTreeClassifier()
model1.fit(X_train, y_train)

print("Model1 DecisionTreeClassifier: ")
print(model1.score(X_test, y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier

model2 = RandomForestClassifier()
model2.fit(X_train, y_train)

print("Model2 RandomForestClassifier: ")
print(model2.score(X_test, y_test))
```

```python
from sklearn.svm import SVC

model3 = SVC()
model3.fit(X_train, y_train)

print("Model3 SVC: ")
print(model3.score(X_test, y_test))
```

```python
# model naive baiyes
# impor pustaka yang dibutuhkan
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.datasets import load_iris # Contoh dataset

# buat dan latih model KNN
# Inisialisasi model KNN dengan nilai K=3 (misal)
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)

# lakukan prediksi pada data uji
y_pred = knn.predict(X_test)

# evaluasi model
# Hitung akurasi model
accuracy = accuracy_score(y_test, y_pred)
print(f'Akurasi model KNN: {accuracy:.4f}')
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report

# model naive baiyes
model_nb = GaussianNB()
model_nb.fit(X_train, y_train) # Melatih model dengan data training

# lakukan prediksi
y_pred = model_nb.predict(X_test)
print("Prediksi Naive Bayes:", y_pred)

# evaluasi model
akurasi = accuracy_score(y_test, y_pred)
print("Akurasi Naive Bayes:", akurasi)

# tampilkan laporan klasifikasi yang lebih detail
print("\nLaporan Klasifikasi:\n", classification_report(y_test, y_pred))
```
